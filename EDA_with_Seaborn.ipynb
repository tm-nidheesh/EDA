Data Preparation

* Dropping irrelevent rows and columns.
* Identifying duplicate columns.  
* Renaming columns.

```python
df.head()
```

```python
df = df[['coaster_name',
    #'Length', 'Speed',
    'Location', 'Status',
    #'Opening date','Type',
    'Manufacturer',
    #'Height restriction', 'Model', 'Height',
    #'Inversions', 'Lift/launch system', 'Cost', 'Trains', 'Park section',
    #'Duration', 'Capacity', 'G-force', 'Designer', 'Max vertical angle',
    #'Drop', 'Soft opening date', 'Fast Lane available', 'Replaced',
    #'Track layout', 'Fastrack available', 'Soft opening date.1',
    #'Closing date', 'Opened',
    #'Replaced by', 'Website',
    #'Flash Pass Available', 'Must transfer from wheelchair', 'Theme',
    #'Single rider line available', 'Restraint Style',
    #'Flash Pass available', 'Acceleration', 'Restraints', 'Name',
       'year_introduced', 'latitude', 'longitude', 'Type_Main',
       'opening_date_clean',
    #'speed1', 'speed2', 'speed1_value', 'speed1_unit',
       'speed_mph',
    #'height_value', 'height_unit',
    'height_ft',
       'Inversions_clean', 'Gforce_clean']].copy()
```

```python
df.shape
```

```python
df.dtypes
```

```python
#Changing the datatype of column

df['opening_date_clean'] = pd.to_datetime(df['opening_date_clean'])
```

```python
#Renaming columns

df = df.rename(columns = {'coaster_name' : 'Coaster_Name',
                     'year_introduced' : 'Year_Introduced',
                     'opening_date_clean' : 'Opening_Date',
                     'speed_mph' : 'Speed_mph',
                     'height_ft' : 'Height_ft',
                     'Inversions_clean' : 'Inversions',
                     'Gforce_clean' : 'Gforce'})
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.columns
```

```python
df = df.loc[~df.duplicated(subset = ['Coaster_Name', 'Location','Opening_Date'])] \
  .reset_index(drop=True).copy()
```

```python
df.shape
```

Understanding Features

Univariate Analysis.

*   Histogram.
*   Kde.
*   Boxplot.


```python
df
```

```python
ax = df['Year_Introduced'].value_counts().head(10).plot(kind = 'bar', title = "Top 10 Years with most Coaster Introduced")
ax.set_xlabel("Year")
ax.set_ylabel("Count")
```

```python
ax = df['Speed_mph'].plot(kind ='hist', bins=20, title = 'Coaster Speed in mph')
ax.set_xlabel("Speed(mph)")
```

```python
ax = df['Speed_mph'].plot(kind ='kde', title = 'Coaster Speed in mph')
ax.set_xlabel("Speed(mph)")
```

Feature Relationships

```python
df.head()
```

```python
df.plot(kind ='scatter', x='Speed_mph', y='Height_ft', title = 'Coaster Speed vs Height')
plt.show()
```

```python
ax = sns.scatterplot(data=df, x='Speed_mph', y='Height_ft', hue='Year_Introduced')
ax.set_title("Speed vs Height")
plt.show()
```

```python
df.head()
```

```python
sns.pairplot(df, vars=['Year_Introduced', 'Speed_mph', 'Height_ft', 'Inversions', 'Gforce'])
plt.show()
```

```python
df_corr = df[['Year_Introduced', 'Speed_mph', 'Height_ft', 'Inversions', 'Gforce']].dropna().corr()
df_corr
```
